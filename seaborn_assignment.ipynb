```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
Seaborn is a powerful data visualization library in Python built on top of Matplotlib.
It provides a higher-level interface for creating a variety of informative and aesthetically pleasing statistical graphics.

Here are five types of plots that can be created using Seaborn, along with their uses:

1. scatter plot:
    Use: Scatter plots display the relationship between two numerical variables.
    They are particularly useful for identifying patterns, correlations, and potential outliers in the data.
    
2. histogram :
    displaying the frequency or count of observations in each bin.
    
3. relational plot:
     relplot provides a flexible and easy-to-use way to create scatter plots and other relational plots with additional features such as color-coding and faceting.
        
4. categorial plot:
    Categorical plots are useful for visualizing how different categories compare or interact with each other.
    
5. joint plot:
    The main plot in a joint plot is usually a scatter plot, and additional plots can include histograms, kernel density estimation (KDE) plots, and regression lines.
    jointplot helps you understand the joint distribution and the individual distributions of the variables simultaneously.
    
    
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


titanic=sns.load_dataset('titanic')

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
sns.boxplot(data=titanic,x='pclass',y='age')
plt.title("Box plot of age bt pclass")

plt.subplot(1,2,2)
sns.boxplot(data=titanic,x='pclass',y='fare')
plt.title("Box plot of fare bt pclass")

plt.tight_layout()

plt.show()
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for a heatmap
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap using Seaborn
plt.figure(figsize=(10, 6))
sns.heatmap(data=flights_pivot, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Flights Passengers Heatmap')
plt.show()
```
